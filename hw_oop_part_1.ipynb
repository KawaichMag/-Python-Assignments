```python
class Encapsulated:
    def __init__(self, value):
        self.__value = None
        self.value = value

    # Your code here (@property & @value.setter)

    @property
    def value(self):
        return self.__value

    @value.setter
    def value(self, new_value):
        if not isinstance(new_value, (int, float)):
            raise TypeError("Wrong Type Assignment (int или float).")

        self.__value = new_value


obj = Encapsulated(10)
print(obj.value)
print(obj._Encapsulated__value)
```

## Задание 3

Реализуйте дженерик-функцию (duck typing) с использованием протоколов (PEP 544) и типовых подсказок, чтобы показать полиморфизм без наследования.

```python
from typing import Protocol


class Drawable(Protocol):
    def draw(self) -> None: ...


# Your code here (create Circle and Square)


class Circle:
    def draw(self) -> None:
        r = 10
        wb_r = 6

        for y in range(wb_r, -wb_r, -1):
            for x in range(-wb_r * 3, wb_r * 3, 1):
                if x**2 + (y * 2.1) ** 2 <= r**2:
                    print("=", end="")
                else:
                    print(" ", end="")
            print()


class Square:
    def draw(self) -> None:
        print("=" * 8)

        for i in range(2):
            print("=" + " " * 6 + "=")

        print("=" * 8)


def paint(obj: Drawable):
    obj.draw()


paint(Circle())
paint(Square())
```

## Задание 8

Реализуйте класс с пользовательскими dunder-методами: new, init, call, del, repr, str, и объясните, как Python вызывает их в цикле жизни объекта.

```python
class LifeCycle:
    # Your code here
    def __new__(cls):
        print(f"We created new live (memory allocation) {cls}")
        # Call new from object
        return super().__new__(cls)

    def __init__(self):
        print("Initialized it..")

    def __call__(self, *args, **kwds):
        print("Called it using LifeCycle(*args, **kwds)")

    def __delete__(self, instance):
        print("Killed it..")
        del instance

    def __repr__(self):
        return "Representation of life cycle :D"

    def __str__(self):
        return "Assume Life is a string..."


obj = LifeCycle()
print(obj)
obj()
print(repr(obj))
del obj
```

## Задание 9

Создайте контекстный менеджер с помощью специальных методов enter и exit, используйте его вместе с классом, в котором присутствуют методы с разграничением прав доступа.

https://habr.com/ru/articles/739326/

```python
class Access:
    # Your code here

    @property
    def secret(self):
        # Try to access secret variable
        try:
            return self._secret
        except AttributeError:
            return "It's a secret"

    def __enter__(self):
        # We create secret variable only after entering the context
        self._secret = 42
        return self.secret

    def __exit__(self, exc_type, exc, tb):
        # Deleting it after leaving the context
        del self._secret


obj = Access()
print(obj.secret)
with obj as secret_value:
    print(secret_value)
print(obj.secret)
```

## Задание 10

Создайте класс, объекты которого могут быть отслежены с помощью слабых ссылок (weakref). Реализуйте систему, которая хранит слабые ссылки на все созданные объекты класса и автоматически удаляет их из списка при уничтожении объектов. Продемонстрируйте это поведение, выводя текущее количество живых объектов.

```python
import weakref
import gc


class TrackedObject:
    _refs = []

    def __init__(self, name):
        self.name = name

        # Create weak ref and put it to class trackable list
        ref = weakref.ref(
            self,
            lambda r: (
                TrackedObject._refs.remove(r) if r in TrackedObject._refs else None
            ),
        )
        TrackedObject._refs.append(ref)

    # Helper function for status rendering
    @classmethod
    def status(cls):
        live = [r() for r in cls._refs if r() is not None]
        print(f"Live: {len(live)}, Names: {[o.name for o in live]}")


a = TrackedObject("A")
b = TrackedObject("B")
c = TrackedObject("C")
TrackedObject.status()

del b
gc.collect()
TrackedObject.status()

del a, c
gc.collect()
TrackedObject.status()
```
